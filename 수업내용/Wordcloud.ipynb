```python
'''
Word Cloud: 자주 등장하는 텍스트를 중요도나 인기도를 고려해 표현한 것. 텍스트 클라우드 라이브러리.
Word Cloud 만들기 위한 과정
1. 자연어 문장에서 키워드를 추출
2. 키워드가 등장한 빈도를 측정(Counter 이용)
3. 앞에서 전처리한 정보와 Wordcloud 라이브러리를 바탕으로 Wordcloud 생성

konlpy: 한국어 형태소 분석기 라이브러리
'''

```

```python
# 라이브러리 불러오기

import matplotlib.pyplot as plt
from wordcloud import WordCloud

from collections import Counter

#문장에서 명사를 추출하는 형태소 분석 라이브러리
from konlpy.tag import Hannanum
```

```python
national_anthem = """
동해물과 백두산이 마르고 닳도록
하느님이 보우하사 우리나라 만세
무궁화 삼천리 화려 강산
대한 사람 대한으로 길이 보전하세
남산 위에 저 소나무 철갑을 두른 듯
바람 서리 불변함은 우리 기상일세
무궁화 삼천리 화려 강산
대한 사람 대한으로 길이 보전하세
가을 하늘 공활한데 높고 구름 없이
밝은 달은 우리 가슴 일편단심일세
무궁화 삼천리 화려 강산
대한 사람 대한으로 길이 보전하세
이 기상과 이 맘으로 충성을 다하여
괴로우나 즐거우나 나라 사랑하세
무궁화 삼천리 화려 강산
대한 사람 대한으로 길이 보전하세
"""
```

```python
hannanum = Hannanum()
nouns = hannanum.nouns(national_anthem)

nouns[:10]
```

```python
#Counter
counter = Counter(nouns)
counter
```

```python
# WordCloud를 이용해 텍스트 구름 만들기
wordcloud = WordCloud(
    font_path="gulim",
    background_color = "white",
    width=1000,
    height=1000,
)

img = wordcloud.generate_from_frequencies(counter) # counter 같은 요소 WordCloud를 만듦.
plt.imshow(img)
```
