```python
# 시각화에 필요한 라이브러리를 불러오기

import seaborn as sns
```

```python
# 꺾은선 그래프 (Lineplot)

sns.lineplot(x = [1,3,2,5], y=[0.7,0.2,0.1,0.05])
```

```python
# 막대 그래프(Barplot) .barplot()
sns.barplot(x=["Amy", "Bob", "Cat", "Dog"], y=[0.7,0.2,0.1,0.05])
```

```python
# matplotlib.pyplot의 속성을 변경해서 그래프에 다양한 요소를 변경/추가 할 수 있음. 
# 불러오기
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=["Amy", "Bob", "Cat", "Dog"], y=[0.7,0.2,0.1,0.05])
plt.title("Bar Plot") # 제목
plt.show()
```

```python
# plt.xlabel, plt.ylabel: 그래프 축에 설명을 추가함. 무게, 길이 등
sns.barplot(x=["Amy", "Bob", "Cat", "Dog"], y=[0.7,0.2,0.1,0.05])
plt.xlabel("X Label")
plt.ylabel("Y Label")
plt.show()
```

```python
# plt.xlim, plt.ylim: 그래프 축의 범위를 지정

sns.lineplot(x=[1,3,2,4], y=[4,3,2,1])
plt.ylim(0,10) #시작위치는 2, 종료위치는 3
plt.show()
```

```python
# plt.figure(figsize = (x,y)): 그래프의 크기를 지정
plt.figure(figsize=(20,10)) # 앞에다 써줘야 함. 
sns.lineplot(x=[1,3,2,4], y=[4,3,2,1])

plt.show()
```
