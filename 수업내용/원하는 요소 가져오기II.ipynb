```python
# User-Agent 추가하기

user_agent = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36"}
```

```python

import requests
from bs4 import BeautifulSoup

# 요청 진행
res = requests.get("https://hashcode.co.kr/", user_agent)
```

```python
soup = BeautifulSoup(res.text, "html.parser")
```

```python
# 질문의 제목을 모아서 가져오기

soup.find("li", "question-list-item").find("div", "question").find("div","top").h4.text #연쇄적
```

```python
questions = soup.find_all("li", "question-list-item")

for question in questions:
    print(question.find("div", "question").find("div", "top").h4.text)
```

```python
# 페이지네이션: 많은 정보를 인덱스로 구분하는 기법. Query String을 통해서 이를 구분할 수 있음. 

# 페이지네이션이 되어있는 질문 리스트의 제목을 모두 가져오기
# 과도한 요청을 방지하기 위해 0.5초마다 요청을 보내보자.

import time

for i in range(1,6):
    res = requests.get("https://hashcode.co.kr/?page={}".format(i), user_agent)
    soup = BeautifulSoup(res.text, "html.parser")
    
    questions = soup.find_all("li", "question-list-item")
    for question in questions:
        print(question.find("div", "question").find("div", "top").h4.text)
    time.sleep(0.5)
```
