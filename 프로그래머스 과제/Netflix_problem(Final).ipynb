```python
import pandas as pd
```

```python
Netflix = pd.read_csv("netflix_titles.csv")
```

```python
Netflix # 넷플릭스 데이터
```

```python
# 한국작품 갯수 구하기

Netflix[Netflix["country"] == "South Korea"] #행의 갯수=한국작품 수
```

```python
from collections import Counter
```

```python
# 제일 많은 작품 수를 가진 국가 구하기

Counter(Netflix["country"]) #미국이 1등(2818개). 한국은 여기서도 199개 인걸 확인할 수 있음.
```

```python
# 제일 많은 작품 수를 가진 국가 구하기 (단일 국가인 경우를 기준으로)
Netflix_country = Netflix.groupby('country')
Netflix_country.count().sort_values(by='show_id', ascending=False) # show_id로 정렬해 줌 + 내림차순
```
