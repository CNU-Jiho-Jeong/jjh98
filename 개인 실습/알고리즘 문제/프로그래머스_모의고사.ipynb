```python
def solution(answers):
    answer = []
    li = []
    number_1 = [1, 2, 3, 4, 5] # 1  
    number_2 = [2, 1, 2, 3, 2, 4, 2, 5] # 2  
    number_3 = [3, 3, 1, 1, 2, 2, 4, 4, 5, 5] # 3  
    count_1 = 0 # 1   
    count_2 = 0 # 2   
    count_3 = 0 # 3   
    k = 0 
    j = 0
    l = 0
    
    for i, element in enumerate(answers):
        if element == number_1[k]:
            count_1 += 1
        k += 1
        if k>=5: #  5 
            k=0
        if element == number_2[j]:
            count_2 += 1
        j += 1
        if j>=8: #  8 
            j=0
        if element == number_3[l]:
            count_3 += 1
        l += 1
        if l>=10: #  10 
            l=0
    
    k = count_1
    j = count_2
    l = count_3
    
    li.append(k)
    li.append(j)
    li.append(l)
    li.sort(reverse=True) #   

    if li[0] != li[1]: #      
        if li[0] == k:
            answer.append(1)
        elif li[0] == j:
            answer.append(2)
        elif li[0] == l:
            answer.append(3)
    elif li[0] == li[1]: #     2 or 3
        if li[2] == k:
            answer.append(2)
            answer.append(3)
            if li[1] == li[2]:
                answer.append(1)
        elif li[2] == j:
            answer.append(1)
            answer.append(3)
            if li[1] == li[2]:
                answer.append(2)
        elif li[2] == l:
            answer.append(1)
            answer.append(2)
            if li[1] == li[2]:
                answer.append(3)
    answer.sort()
    return answer
```
