```python
# 첫 번째 풀이방법 (재귀함수 방식)

n = int(input())
def sum(n):
    if n<=1:
        return n
    else:
        return n + sum(n-1)
print(sum(n))
```

## 첫 번째 풀이방법

### (1) 문제번호: 8393 합
### (2) 사용 알고리즘: 재귀알고리즘
### (3) 문제의 실마리: "1 부터 n까지" -> 차례대로!
### (4) 시간복잡도: O(n) (런타임 에러 발생)
### (5) 공간복잡도: 73699 KB

```python
# 두 번째 풀이방법 (for문 활용)

n = int(input())
a = 0
for i in range(1, n+1):
    a += i
print(a)
```

## 두 번째 풀이방법

### (1) 문제번호: 8393 합
### (2) 사용 알고리즘: 선형 시간 알고리즘
### (3) 문제의 실마리: "1 부터 n까지" -> 차례대로!
### (4) 시간복잡도: O(n) (정답)
### (5) 공간복잡도: 30840 KB

```python
# 세 번째 풀이방법 (분할법 활용)

N = int(input())
def sum_for_divide_conquer(N):
    def generate(n):
        if n == 1:
            return 1
        elif n % 2 == 1:
            return n + generate(n-1)
        else:
            return generate(n // 2) * 2 + (n ** 2) // 4
    return generate(N)
print(sum_for_divide_conquer(N))
```

## 세 번째 풀이방법



### (1) 문제번호: 8393 합
### (2) 사용 알고리즘: 로그 시간 알고리즘 
### (3) 문제의 실마리: "1 부터 n까지" -> 차례대로!
### (4) 시간복잡도: O(log2n) (정답) (입력을 절반으로 줄여나가면서 1까지 만듦)
### (5) 공간복잡도: 30840 KB

```python
# 네 번째 풀이방법 (가우스 공식)

n = int(input())
print(int((n+1)*n/2))
```

## 네 번째 풀이방법

### (1) 문제번호: 8393 합
### (2) 사용 알고리즘:  O(1) (즉각적인 결과)
### (3) 문제의 실마리: "1 부터 n까지" -> 차례대로!
### (4) 시간복잡도: O(1) (정답)
### (5) 공간복잡도: 30840 KB
